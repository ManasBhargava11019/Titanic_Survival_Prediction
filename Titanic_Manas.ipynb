```python
train = train.drop(['Name'],axis =1)
```

```python
train.head()
```

```python
train.Cabin= train.Cabin.fillna('0')
train.Embarked=train.Embarked.ffill()
train.head()
```

```python
train.loc[train['Cabin']!="0",'Cabin'] =1
train.Cabin=train.Cabin.astype(str).astype(int)
```

```python
features = ["Pclass","Sex", "SibSp", "Parch","Embarked"]
train_feat = pd.get_dummies(train[features])
train.head()
```

```python
train = train.drop(columns=(["Pclass","Sex","SibSp","Parch","Ticket","Embarked"]),axis =1)
train.head()
```

```python
train.Age = train.Age.fillna(train.Age.mean())
train.isnull().sum()
```

```python
data = [train,train_feat]
train = pd.concat(data,axis=1,join='inner')
train
train.isnull().sum()
```

```python
test = test.drop(['Name'],axis=1)
test.Cabin= test.Cabin.fillna('0')
test.loc[test['Cabin']!="0",'Cabin'] =1
test.Cabin=test.Cabin.astype(str).astype(int)
test
```

```python
test['Age'] = test['Age'].fillna(test['Age'].mean())
test['Fare'] = test['Fare'].fillna(test['Fare'].mean())
test.isnull().sum()
#test.Cabin.dtype
```

```python

features = ["Pclass","Sex", "SibSp", "Parch","Embarked"]
test_feat = pd.get_dummies(test[features])
test=test.drop(columns=(["Pclass","Sex","SibSp","Parch","Ticket","Embarked"]),axis =1)
data = [test,test_feat]
test = pd.concat(data,axis=1,join='inner')
```
